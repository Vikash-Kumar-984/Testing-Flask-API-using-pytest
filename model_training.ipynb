```python
#Splitting the dataset into training and testing
y= df['bwt']
```

```python
x=df.iloc[:,1:]
x.head()
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split as tts

x_train, x_test, y_train, y_test = tts(x,y, test_size=0.20)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge

#Creating instance of the model
lr= LinearRegression()
lasso = Lasso()
ridge = Ridge()


#Training the model with training dataset
lr.fit(x_train,y_train)
lasso.fit(x_train,y_train)
ridge.fit(x_train,y_train)
```

```python
#Predicting the model with testing data
lr_pred=lr.predict(x_test)
las_pred = lasso.predict(x_test)
rid_pred = ridge.predict(x_test)
```

```python
# Model Evaluation

from sklearn.metrics import r2_score, mean_squared_error

#R2_Score Model
lr_rscore =  r2_score(y_test, lr_pred)
las_rscore =  r2_score(y_test, las_pred)
rid_rscore =  r2_score(y_test, rid_pred)
```

```python
#mean_squared_error Model
lr_mse =  mean_squared_error(y_test, lr_pred)
las_mse =  mean_squared_error(y_test, las_pred)
rid_mse =  mean_squared_error(y_test, rid_pred)
```

```python
print("\n The performance of the Linear Regression Model is: ")
print("R2 Score : ", lr_rscore)
print("MSE: ",lr_mse)

print("\n The performance of the Lasso Regression Model is: ")
print("R2 Score : ", las_rscore)
print("MSE: ",las_mse)

print("\n The performance of the Ridge Regression Model is: ")
print("R2 Score : ", rid_rscore)
print("MSE: ",rid_mse)

```
